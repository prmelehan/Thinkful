# Wage difference investigation

A brief look into the Wage breakdown at different levels.
I'm curious because it might give some insight into exactly what's going on with attrition

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap

%matplotlib inline
```

```python
employee_survey_data = pd.read_csv("../data/hr/employee_survey_data.csv")
general_data = pd.read_csv("../data/hr/general_data.csv")
manager_survey_data = pd.read_csv("../data/hr/manager_survey_data.csv")
```

```python
employee_survey_data = employee_survey_data.set_index("EmployeeID")
general_data = general_data.set_index("EmployeeID")
manager_survey_data = manager_survey_data.set_index("EmployeeID")
```

```python
# make a full df which merges on the index
full_df = pd.concat([employee_survey_data, general_data, manager_survey_data], axis="columns")
```

### Breakdown Income: Job Level

```python
full_df.JobLevel.value_counts(dropna=False)
```

```python
for job_level in list(sorted(set(full_df.JobLevel))):
    # print the distribution of monthly income for this job level
    monthly_income = full_df[full_df.JobLevel == job_level]["MonthlyIncome"]
    
    sns.distplot(monthly_income, kde=False)
```

Seems like there is **not** a large different in income per job level.

```python
sns.scatterplot(full_df.JobLevel, full_df.MonthlyIncome)
```
